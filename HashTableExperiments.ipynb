```python
from openAddressing import DoubleProbingHashST
```

### Initial table size = 11

```python
d = DoubleProbingHashST(M=11)
with open("TaleOfTwoCities.txt", mode='r', encoding='utf-8') as f:
    for line in f:
        words = (word for word in line.split() if len(word) > 7)
        for w in words:
            if d.get(w):
                d.put(w, d.get(w) + 1)
            else:
                d.put(w, 1)
d.a.plotData(xlim=50, ylim=12)
```

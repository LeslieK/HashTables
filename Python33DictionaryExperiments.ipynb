## Experiment: Insert words of "Tale Of Two Cities" into a CPython 3.3 Hash Table (dict data structure)

```python
%matplotlib inline
```

```python
from visualizerAccumulator import VisualAnalyzer
from Py33DictExperiments import pythonDict, countProbes
```

```python
a = VisualAnalyzer(13525, 12, title="Python 3.3 dictionary")
pmap = pythonDict()
probes = countProbes(pmap)
```

```python
number_of_keys = len(pmap)
print('number of keys: ', number_of_keys)
```

```python
# plot data
with open("TaleOfTwoCities.txt", mode='r', encoding='utf-8') as f:
    words = (word for line in f for word in line.split() if len(word) > 7)
    for key in words:
        number_of_probes = probes[key]
        a.addDataValue(number_of_probes)
```

```python
a.plotData(xlim=100, ylim=4)
```
